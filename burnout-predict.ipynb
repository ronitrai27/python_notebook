```python
# (Label Encoding)
df["Experience_Level"] = df["Experience_Level"].map({
    "Junior": 0,
    "Mid-Level": 1,
    "Senior": 2,
    "Tech Lead": 3
})  

df["Burnout_Risk"] = df["Burnout_Risk"].map({
    "Low":0,
    "Medium":1,
    "High":2
})

df.head()

```
