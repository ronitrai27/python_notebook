```python
import pandas as pd

df = pd.read_csv("data/NIRF Ranking 2025.csv")

df.head()
```

```python
df.info()
df.columns
df.shape
```

```python
# ques 1: Which college is ranked 1st in the NIRF Ranking 2025?
# ques 2: Show the top 5 institutes based on Score.
# ques 3: Find the average Score of all institutes.
#  ques 4: List all institutes located in "Delhi".
# ques 5: Show Institute Name, City, and Rank columns only.
# ques 6 (bonus): Count how many institutes exist in each State.
# ques 7 (bonus): Find the institute with the highest PERCEPTION score.
```

```python
# ================================
# Q1: Which college is ranked 1st?
# ================================

# Filter rows where Rank == 1
rank1_college = df[df["Rank"] == 1]

print("\nQ1: College ranked 1st in NIRF 2025:\n")
print(rank1_college)


# =========================================
# Q2: Show the top 5 institutes by Score
# =========================================

# Sort by Score in descending order and take first 5 rows
top5_score = df.sort_values("Score", ascending=False).head(5)

print("\nQ2: Top 5 institutes based on Score:\n")
print(top5_score)


# =========================================
# Q3: Find the average Score of all institutes
# =========================================

# Calculate mean of Score column
avg_score = df["Score"].mean()

print("\nQ3: Average Score of all institutes:\n")
print(avg_score)


# =========================================
# Q4: List all institutes located in Delhi
# =========================================

# Filter where City == Delhi
delhi_institutes = df[df["City"] == "Delhi"]

print("\nQ4: Institutes located in Delhi:\n")
print(delhi_institutes)


# =========================================
# Q5: Show Institute Name, City, and Rank only
# =========================================

# Select specific columns
selected_columns = df[["Institute Name", "City", "Rank"]]

print("\nQ5: Institute Name, City, and Rank columns:\n")
print(selected_columns)


# =========================================
# Q6 (BONUS): Count institutes in each State
# =========================================

# Count values in State column
state_counts = df["State"].value_counts()

print("\nQ6: Number of institutes in each State:\n")
print(state_counts)


# =========================================
# Q7 (BONUS): Institute with highest PERCEPTION score
# =========================================

# Find index of max PERCEPTION value
max_perception_index = df["PERCEPTION"].idxmax()

# Get full row of that institute
highest_perception = df.loc[max_perception_index]

print("\nQ7: Institute with highest PERCEPTION score:\n")
print(highest_perception)
```
