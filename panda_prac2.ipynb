```python
# Pnadas dataframe creation
import pandas as pd
import numpy as np

# Create a dictionary of data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve','Missing', 'Sameer'],
    'Age': [25, 23, 'Missing', 22, 25, 24, 26],
    'Marks': [85, np.nan, np.nan, 96, 88, 90, 82],
    'Subject': ['Math', 'Physics', 'Math', 'Chemistry', 'Math', 'Physics', 'Chemistry']
}

```

```python
# check shape
# diplay data types of columns
# generate statistics of numerical columns (min/max/mean)

# sort dataframe by Marks in decending order
# use specific pandas function to find the top 2 highest scores (scholarship candidates) without sorting the whole dataframe.
# identify the student with lowest marks (remedial candidate)
```

```python
df = pd.DataFrame(data)
print("DataFrame Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nStatistics of Numerical Columns:\n", df.describe())

# Sort by Marks in descending order
sorted_df = df.sort_values(by='Marks', ascending=False)
print("\nDataFrame sorted by Marks (descending):\n", sorted_df)
# Find top 2 highest scores without sorting the whole dataframe
top_2_scores = df.nlargest(2, 'Marks')
print("\nTop 2 highest scores without sorting the whole dataframe:\n", top_2_scores)

# Identify the student with lowest marks (remedial candidate)
lowest_score = df.nsmallest(1, 'Marks')
print("\nStudent with the lowest marks (remedial candidate):\n", lowest_score)
```
