```python
import pandas as pd
import numpy as np 
import warnings
warnings.filterwarnings('ignore')

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', np.nan],
    'Age': [25, 30, np.nan, 40, 35, 28],
    'Salary': [50000, 60000, 55000, np.nan, 70000, 65000],
    'City': ['New York', 'New York', 'Chicago', np.nan, 'Houston', 'Chicago'],
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT', 'Finance'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Female'],
    'joiningDate': ['2020-01-15', '2019/03/10', '2021-06-20', '2018-11-05', '2020-07-30', '2019-12-01'],
    'wifeName': ['sarina', 'Alice', np.nan, 'Eve', np.nan, 'mia'],
    'wifeAge': [28, 25, np.nan, 30, np.nan, 24]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Data Ingestion (Loading Data)
# Ingestion means reading data from various sources into a pandas DataFrame.
# pd.read_csv('file.csv'): Load CSV files. Use sep=',' for custom delimiters
# pd.read_excel('file.xlsx'): Load Excel files.
# pd.read_json('file.json'): Load JSON data.
```

```python
df.head()
df.tail()
print("\info :\n")
df.info()
df.shape
df.describe()
df.select_dtypes(include="number")
df.select_dtypes(include="object")
df.isnull().sum()
df["City"].value_counts()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
df["City"] = df["City"].fillna(df["City"].mode()[0])

print("\nAfter handling missing values:\n")
print(df)
```

```python

# df.dropna(axis=1)   # drop columns
# df.drop_duplicates()
# df.rename(columns={"old": "new"}, inplace=True)
df["Age"] = df["Age"].astype(int)
# df["joiningDate"] = pd.to_datetime(df["joiningDate"])
df["joiningDate"] = pd.to_datetime(df["joiningDate"], errors="coerce")
print("\nAfter type conversion:\n")
print(df)
# df.dropna()
# print("\nAfter dropping rows with missing values:\n")
# print(df)

```
